```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
plt.plot(x,y)
plt.show()
plt.plot(x,y,marker='o')
plt.show()
```

```python
plt.plot(x,y,linestyle='dashed')
plt.show()
plt.plot(x,y,'hotpink')
plt.show()
plt.plot(x,y,linewidth='20',color='r')
plt.grid()
plt.show()
plt.scatter(x,y)
plt.show()
```

```python

import numpy as np
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x,y)
plt.show()
```
